```python
from plotnine import ggplot, aes, geom_line # import the ggplot() class as well as some useful functions from plotnine, aes() and geom_line()
# A plot of population over time
(
    ggplot(economics) # What data to use
    + aes(x="date", y="pop") # What variable to use
    + geom_line() # Geometric object to use for drawing
)
```

 ## Staistical Transformations: Aggregate and Transform your Data
 
 ### Create a histogram to display distributions of Lake Huron from 1875 to 1975

```python
from plotnine.data import huron # dataset with the levels of Lake Huron 1875-1975

print(huron)

from plotnine import ggplot, aes, stat_bin, geom_bar # stat_bin() divides the level range into 20 equally sized bins. Then the number of measurements that falls into each bin is drawn using a bar plot.

ggplot(huron) + aes(x="level") + stat_bin(bins = 20) + geom_bar() # geom_histogram(bins=20) can also be used
```

```python
from plotnine import ggplot, aes, geom_boxplot

print(ggplot(huron) 
      + aes(x="factor(decade)", y="level")  
      + geom_boxplot()
     )
```
